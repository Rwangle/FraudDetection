```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
import numpy as np
import gc # Added for memory management


df = dfTrainID.copy()

# 1. Drop extreme missing cols and low variance cols
high_missing = [col for col in df.columns if df[col].isna().mean() > 0.95]
# Identify columns where one value makes up >90% of the data (low variance)
big_top_value_cols = [col for col in df.columns if df[col].value_counts(dropna=False, normalize=True).values[0] > 0.9]

cols_to_drop = list(set(high_missing + big_top_value_cols))

df.drop(columns=cols_to_drop, inplace=True)
print(f"Dropped {len(cols_to_drop)} columns due to high missingness or low variance.")


# 2. Data for Random Forest (requires full processing)
df_rf = df.copy()
# Impute numeric
num_cols = df_rf.select_dtypes(include=['float64', 'int64']).columns
imputer_num = SimpleImputer(strategy='median')
df_rf[num_cols] = imputer_num.fit_transform(df_rf[num_cols])

# Impute and encode categorical using OrdinalEncoder with unknown handling
cat_cols = df_rf.select_dtypes(include=['object']).columns

# Use SimpleImputer with 'constant' strategy to fill NaNs with "Missing" as a category
imputer_cat = SimpleImputer(strategy='constant', fill_value="Missing")
df_rf[cat_cols] = imputer_cat.fit_transform(df_rf[cat_cols])

# Apply OrdinalEncoder (would be a problem due to weight in linear but works fine in tree based)
encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
df_rf[cat_cols] = encoder.fit_transform(df_rf[cat_cols])


# 3. Data for XGBoost / CatBoost (minimal processing, let them handle missing values)
df_native = df.copy() 
# Ensure missing values are represented as np.nan (default)
# For CatBoost, convert categorical columns to 'category' dtype if using native handling
for col in df_native.select_dtypes(include=['object']).columns:
    df_native[col] = df_native[col].astype('category')
df_native.shape

#gc.collect()
```
